```python
import sklearn.feature_extraction.text as sk_text
from sklearn.datasets import fetch_20newsgroups

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = fetch_20newsgroups()
```

```python
type(data)
```

```python
data.target_names
```

```python
categories = [
    'sci.electronics',
    'sci.space',
     'comp.graphics',
    'rec.motorcycles',
]
```

```python
print(data.data[0])
```

```python
data = fetch_20newsgroups(categories = categories)
```

```python
data.target_names
```

```python
data.data
```

```python
cv = sk_text.CountVectorizer(stop_words="english")
```

```python
X = cv.fit_transform(data.data).toarray()
```

```python
len(X[0])
```

```python
len(data.data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, data.target, test_size=0.3)
```

```python
nb = MultinomialNB()
```

```python
nb.fit(X_train, y_train)
```

```python
y_pred = nb.predict(X_test)
```

```python
data.target_names
```

```python
report_cv = classification_report(y_test, y_pred, target_names=data.target_names)
```

```python
print(report_cv)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, data.target, test_size=0.3)
nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)
report_cv = classification_report(y_test, y_pred, target_names=data.target_names)
print(report_cv)
```

```python
confusion_matrix(y_test, y_pred)
```
